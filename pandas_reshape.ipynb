```python
'''
First drop TeamName from table since it is not shown in the final wide table
Add Year, Player and Team to Index since we only want player stats to be stacked into a series
Stack game stats into a PD series
In order to reobtain the column labels in the right order, we need to "pop" or unstack in a certain order

By indices:
0 = Year
1 = Player
2 = Team
3 = variable (player stats)

The order of bball_wide is Player, variable, then Team for column labels, so the unstack order must be (by index) 
a list [1,3,2] to create all the column labels we need, keeping year as the index of the dataframe

Finally sort the data to obtain bball_wide
'''
bball.drop("TeamName", axis=1).set_index(["Year", "Player", "Team"]).stack().unstack(level=[1,3,2]).sort_index(axis=1)
```
